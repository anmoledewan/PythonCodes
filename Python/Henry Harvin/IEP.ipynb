```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 5
```

```python
data = pd.read_csv('index general maharashtra.csv')
print(data.head())
print('\n Data Types:')
print(data.dtypes)
data.index
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%m-%Y')
data = pd.read_csv('index general maharashtra.csv', parse_dates=['Month'], index_col='Month',date_parser=dateparse)
print(data.head())
print(data.dtypes)
```

```python
data.index
```

```python
ts = data['Rural']
```

```python
fig = plt.figure()
plt.plot(pd.DataFrame(data,columns=["Rural","Urban"]))
fig.suptitle('CPI - General - Maharashtra', fontsize=15)
plt.gca().legend(('Rural','Urban'))
plt.show()
fig.savefig('magGen.jpg')
```
