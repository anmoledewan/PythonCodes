```python


file_reader_Old = pd.read_csv(r'L2018143.csv', header=None)
file_reader_Old.columns = ["COCODE", "LINE_NO", "QUARTER", "GRP_CODE","GRP_NM","NAIC_COMPANY_CODE","ID_NBR","FED_RSSD","CIK","NM_OF_SECURITIES_EXCHANGE_IF_P","NM_OF_PARENT_SUBSIDIARIES_OR_A","DOM_LOCATION","RELATIONSHIP_TO_REPORTING_ENTI","DIRECTLY_CNTRL_BY","TYPE_OF_CONTROL","IF_CONTROL_IS_OWNERSHIP_PROVID","ULTIMATE_CONTROLLING_ENTITIES","IS_AN_SCA_FILING_REQUIRED","ASTERISK","LEI"]
file_reader_Old=file_reader_Old.dropna()
file_reader_New = pd.read_csv(r'L18Q3022.csv', header=None)
file_reader_New.columns = ["COCODE", "LINE_NO", "QUARTER", "GRP_CODE","GRP_NM","NAIC_COMPANY_CODE","ID_NBR","FED_RSSD","CIK","NM_OF_SECURITIES_EXCHANGE_IF_P","NM_OF_PARENT_SUBSIDIARIES_OR_A","DOM_LOCATION","RELATIONSHIP_TO_REPORTING_ENTI","DIRECTLY_CNTRL_BY","TYPE_OF_CONTROL","IF_CONTROL_IS_OWNERSHIP_PROVID","ULTIMATE_CONTROLLING_ENTITIES","IS_AN_SCA_FILING_REQUIRED","ASTERISK","LEI"]
file_reader_New=file_reader_New.dropna()
```

```python
file_reader_Old=file_reader_Old[['COCODE','NAIC_COMPANY_CODE','ID_NBR','NM_OF_SECURITIES_EXCHANGE_IF_P',
                                 'NM_OF_PARENT_SUBSIDIARIES_OR_A','RELATIONSHIP_TO_REPORTING_ENTI','DIRECTLY_CNTRL_BY',
                                'TYPE_OF_CONTROL','IF_CONTROL_IS_OWNERSHIP_PROVID',
                                 'ULTIMATE_CONTROLLING_ENTITIES']]

file_reader_New=file_reader_New[['COCODE','NAIC_COMPANY_CODE','ID_NBR','NM_OF_SECURITIES_EXCHANGE_IF_P',
                                 'NM_OF_PARENT_SUBSIDIARIES_OR_A','RELATIONSHIP_TO_REPORTING_ENTI','DIRECTLY_CNTRL_BY',
                                'TYPE_OF_CONTROL','IF_CONTROL_IS_OWNERSHIP_PROVID',
                                 'ULTIMATE_CONTROLLING_ENTITIES']]
```

```python
#Concatenated columns from first Dataframe 
df = file_reader_Old
df['Key'] =  df['NM_OF_PARENT_SUBSIDIARIES_OR_A'].map(str)+ ' ' + df['DIRECTLY_CNTRL_BY']
df['Age']="N"
file_reader_Old= df

df = file_reader_New
df['Key'] =  df['NM_OF_PARENT_SUBSIDIARIES_OR_A'].map(str)+ ' ' + df['DIRECTLY_CNTRL_BY']
df['Age']="N"
file_reader_New=df

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('old.xlsx', engine='xlsxwriter')
# Convert the dataframe to an XlsxWriter Excel object.
file_reader_Old.to_excel(writer, sheet_name='Sheet1')
# Close the Pandas Excel writer and output the Excel file.
writer.save()

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('new.xlsx', engine='xlsxwriter')
# Convert the dataframe to an XlsxWriter Excel object.
file_reader_New.to_excel(writer, sheet_name='Sheet1')
# Close the Pandas Excel writer and output the Excel file.
writer.save()

del file_reader_Old,file_reader_New

```
